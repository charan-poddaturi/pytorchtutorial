```python
x = torch.tensor(6.8)
y = torch.tensor(0.0)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def binary_crossentropy_loss(prediction,target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction,epsilon,1-epsilon)
    return - (target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_crossentropy_loss(y_pred,y)
```

```python
loss
```

```python
dloss_dy_pred = (y_pred - y) / (y_pred * (1 - y_pred))
dy_pred_dz = y_pred * (1 - y_pred)
dz_dw = x
dz_db = 1

dloss_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dloss_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"Manual gradients of loss wrt weight(dw) and wrt bias(db): {dloss_dw}, {dloss_db}")
```
